# Visualizing  of a dataset using seaborn library

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="/content/Automobile (1).csv"
```

```python
data=pd.read_csv(path)
```

```python
data.head()
```

## Plotting univariate distributions

```python
sns.displot(data['normalized_losses'])
```

```python
sns.distplot(data['normalized_losses'])
```

```python
sns.distplot(data['normalized_losses'],hist=False,color="red")
```

```python
sns.distplot(data['normalized_losses'],kde=False,color="red",bins=6,rug=True)
```

```python
sns.distplot(data["curb_weight"],kde=False,color="green",bins=6,rug=True)
```

## Plotting bivariate distributions

SCATTERPLOT

```python
sns.scatterplot(x="horsepower",y="city_mpg",data=data)
```

```python
sns.scatterplot(x="horsepower",y="engine_size",data=data)
```

```python
sns.scatterplot(x="horsepower",y="engine_size",data=data,hue='fuel_system')
```

jointplot

```python
sns.jointplot(data['horsepower'],data["city_mpg"])
```

kind=hex

```python
sns.jointplot(data['horsepower'],data["city_mpg"],kind="hex",color="red")
```

```python
sns.jointplot(data['horsepower'],data["city_mpg"],kind="kde",color="red")
```

```python
sns.jointplot(data['horsepower'],data["city_mpg"],kind="reg",color="green")
```

```python
sns.jointplot(data['horsepower'],data["city_mpg"],kind="resid",color="brown",)
```

Visualizing pairwise relationships in a dataset

```python
sns.pairplot(data[["length","width","height"]])
```

### Plotting with categorical data

```python
sns.set(color_codes=True)
```

```python
sns.stripplot(data["fuel_type"],data["highway_mpg"])
```

```python
data.head()
```

```python
sns.stripplot(data["fuel_type"],data["horsepower"],jitter=True)
```

```python
sns.swarmplot(data["fuel_type"],data["horsepower"])
```

## Boxplots

```python
sns.boxplot(data["engine_type"],data["horsepower"],hue=data["fuel_type"])
```

violinplot

```python
sns.violinplot(data["engine_type"],data["horsepower"],hue=data["fuel_type"])
```

barplot

```python
sns.barplot(data['body_style'], data['horsepower'], hue=data['engine_location'])
```

countplot

```python
sns.countplot(data['body_style'])
```

point plot

```python
sns.pointplot(data["fuel_system"],data["price"],hue=data['number_of_doors'])
```

## Function to draw linear regression models

```python
sns.lmplot(x="horsepower", y="peak_rpm", data=data)
```

```python
sns.lmplot(x="horsepower", y="peak_rpm",hue="fuel_type", data=data)
```
